```python
import matplotlib.pyplot as plt
import numpy as np

# -----------------------------
# Hardcoded model results
# -----------------------------
models = [
    "MLP (3-layer NN)",
    "kNN (k=25)",
    "SVM (Linear)",
    "Logistic Regression",
    "Decision Tree",
    "Random Forest (PCA, Baseline)",
    "Random Forest (PCA, Tuned)",
    "XGBoost (PCA, Baseline)",
    "XGBoost (PCA, Tuned)"
]

accuracy = [
    0.7957, 0.7810, 0.7782, 0.7077, 0.7597,
    0.7871, 0.7777, 0.7569, 0.7465
]

precision = [
    0.7634, 0.5897, 0.5991, 0.4702, 0.5374,
    0.6429, 0.5673, 0.5313, 0.5154
]

recall = [
    0.8566, 0.5750, 0.4964, 0.8179, 0.6679,
    0.4484, 0.6904, 0.7260, 0.7758
]

f1_score = [
    0.8073, 0.5823, 0.5430, 0.5971, 0.5955,
    0.5283, 0.6228, 0.6135, 0.6193
]

roc_auc = [
    0.8695, 0.8133, 0.8084, 0.8266, 0.7978,
    0.8219, 0.8414, 0.8241, 0.8402
]

x = np.arange(len(models))

# -----------------------------
# Figure 1: Accuracy & ROC-AUC
# -----------------------------
plt.figure(figsize=(10, 5))
plt.bar(x - 0.2, accuracy, 0.4, label="Accuracy")
plt.bar(x + 0.2, roc_auc, 0.4, label="ROC-AUC")

plt.xticks(x, models, rotation=30, ha="right")
plt.ylabel("Score")
plt.title("Overall Model Performance (Accuracy vs ROC-AUC)")
plt.legend()
plt.grid(axis="y", alpha=0.3)
plt.tight_layout()
plt.show()

# -----------------------------
# Figure 2: Precision vs Recall
# -----------------------------
plt.figure(figsize=(10, 5))
plt.plot(models, precision, marker="o", label="Precision")
plt.plot(models, recall, marker="o", label="Recall")

plt.xticks(rotation=30, ha="right")
plt.ylabel("Score")
plt.title("PrecisionRecall Tradeoff Across Models")
plt.legend()
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()

# -----------------------------
# Figure 3: F1-score Comparison
# -----------------------------
plt.figure(figsize=(8, 5))
plt.barh(models, f1_score)
plt.xlabel("F1-score")
plt.title("F1-score Comparison Across Models")
plt.grid(axis="x", alpha=0.3)
plt.tight_layout()
plt.show()
```
